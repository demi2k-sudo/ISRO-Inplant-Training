```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def Load(file):
    B1 = pd.read_csv(file)
    B1.columns = ['DATE','COUNT']
    return B1
```

```python
B1_S = Load('BSNL1_SUCCESS.csv')
B1_F = Load('BSNL1_FAILURE.csv')
B2_S = Load('BSNL2_SUCCESS.csv')
B2_F = Load('BSNL2_FAILURE.csv')
BL_S = Load('BSNL_L_SUCCESS.csv')
BL_F = Load('BSNL_L_FAILURE.csv')
DNS_S = Load('DNS_SUCCESS.csv')
DNS_F = Load('DNS_FAILURE.csv')
INT_S = Load('INTRANET_SUCCESS.csv')
INT_F = Load('INTRANET_FAILURE.csv')
NKN1_S = Load('NKN_SUCCESS.csv')
NKN1_F = Load('NKN_FAILURE.csv')
NKN2_S = Load('NKN2_SUCCESS.csv')
NKN2_F = Load('NKN2_FAILURE.csv')
PFMS_S = Load('PFMS_SUCCESS.csv')
PFMS_F = Load('PFMS_FAILURE.csv')
```

```python
def convert(df):
    return df.split('-')
```

```python
convert('2004-09-04')
```

```python
def splitConvert(X):
    X['DATE']=X['DATE'].apply(convert)
    return X
```

```python
B1_S = splitConvert(B1_S)
B1_F = splitConvert(B1_F)
B2_S = splitConvert(B2_S)
B2_F = splitConvert(B2_F)
BL_S = splitConvert(BL_S)
BL_F = splitConvert(BL_F)
DNS_S = splitConvert(DNS_S)
DNS_F = splitConvert(DNS_F)
INT_S = splitConvert(INT_S)
INT_F = splitConvert(INT_F)
NKN1_S = splitConvert(NKN1_S)
NKN1_F = splitConvert(NKN1_F)
NKN2_S = splitConvert(NKN2_S)
NKN2_F = splitConvert(NKN2_F)
PFMS_S = splitConvert(PFMS_S)
PFMS_F = splitConvert(PFMS_F)
PFMS_F
```

```python
def get_month(dt):
    return dt[1]
def get_year(dt):
    return dt[0]
def get_day(dt):
    return dt[2]
```

```python
def splitFrame(X):
    for i in X['DATE']:
        X['Year']=X['DATE'].apply(get_year)
        X['Month']=X['DATE'].apply(get_month)
        X['Day']=X['DATE'].apply(get_day)
    return X
```

```python
B1_S = splitFrame(B1_S)
B1_F = splitFrame(B1_F)
B2_S = splitFrame(B2_S)
B2_F = splitFrame(B2_F)
BL_S = splitFrame(BL_S)
BL_F = splitFrame(BL_F)
DNS_S = splitFrame(DNS_S)
DNS_F = splitFrame(DNS_F)
INT_S = splitFrame(INT_S)
INT_F = splitFrame(INT_F)
NKN1_S = splitFrame(NKN1_S)
NKN1_F = splitFrame(NKN1_F)
NKN2_S = splitFrame(NKN2_S)
NKN2_F = splitFrame(NKN2_F)
PFMS_S = splitFrame(PFMS_S)
PFMS_F = splitFrame(PFMS_F)
```

```python
def showplot(F,mon,yr,st,st1):
    day={}
    il=[]
    temp=0
    cl=[]
    for i in range(1,32):
        day[i]=0
    for i,j,c,y in zip(F['Day'],F['Month'],F['COUNT'],F['Year']):
        if j == mon and y == yr:
            day[int(i)]+=c
            temp+=1
            il.append(i)
            cl.append(c)
    if temp == 0:
        print("No records Found :(")
    else:
        plt.bar(il,cl,color='green')
        plt.xlabel("DATE")
        plt.ylabel(st1.upper()+" COUNT") 
        plt.title(st.upper()+" "+st1.upper())
        plt.yticks(range(0,1440,200))
        plt.show()
        plt.plot(list(day.keys()),list(day.values()),color="red")
        plt.xlabel("DATE")
        plt.ylabel(st1.upper()+" COUNT")
        plt.xticks(range(1,32,1))
        plt.yticks(range(0,1440,200))
        plt.title(st.upper()+" "+st1.upper())
        plt.show()
```
